```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
#Importing dataset

df = pd.read_csv('breast-cancer.csv')
print(df.head())
```

```python
#Drop id and unnameed:32

df.drop(columns=['id'],inplace=True)
#encoding diagnosis:M=1,b=0
le=LabelEncoder()
df['diagnosis']=le.fit_transform(df['diagnosis'])
df.head()

#encoding diagnosis:M=1,b=0
le=LabelEncoder()
df['diagnosis']=le.fit_transform(df['diagnosis'])
df.head()
```

```python
#split sata into training and testing data
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['diagnosis']),df['diagnosis'],test_size=0.2,random_state=42)
#feature scaling
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
#2.Train an SVM with linear and RBF kernel.
svm_linear=SVC(kernel='linear')
svm_linear.fit(X_train,y_train)
svm_rbf=SVC(kernel='rbf')
svm_rbf.fit(X_train,y_train)

svm_linear_pred=svm_linear.predict(X_test)
svm_rbf_pred=svm_rbf.predict(X_test)
svm_linear_acc=accuracy_score(y_test,svm_linear_pred)
svm_rbf_acc=accuracy_score(y_test,svm_rbf_pred)

print(f'Accuracy of SVM with linear kernel:{svm_linear_acc*100:.2f}')
print(f'Accuracy of SVM with RBF kernel:{svm_rbf_acc*100:.2f}')

print(classification_report(y_test,svm_linear_pred))
print(classification_report(y_test,svm_rbf_pred))
```

```python
from sklearn.decomposition import PCA

def plot_decision_boundary(model, X, y, title):
    # Apply PCA to reduce dimensions to 2 for visualization
    pca = PCA(n_components=2)
    X_pca = pca.fit_transform(X)

    # Re-train the model on the 2 PCA components for visualization purposes
    # This is necessary because the original model was trained on 30 features
    model_2d = model.__class__(kernel=model.kernel)
    model_2d.fit(X_pca, y)

    h = .02
    x_min, x_max = X_pca[:, 0].min() - 1, X_pca[:, 0].max() + 1
    y_min, y_max = X_pca[:, 1].min() - 1, X_pca[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    # Predict on the grid using the model trained on 2 PCA components
    Z = model_2d.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(6, 5))
    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, edgecolors='k')
    plt.title(title)
    plt.xlabel("Principal Component 1") # Add labels for clarity
    plt.ylabel("Principal Component 2") # Add labels for clarity
    plt.show()

# Now call the function with the original X_test and y_test
# The function will handle PCA internally for plotting
plot_decision_boundary(svm_linear, X_test, y_test, "SVM with Linear Kernel (PCA Reduced)")
plot_decision_boundary(svm_rbf, X_test, y_test, "SVM with RBF Kernel (PCA Reduced)")
```

```python
# Step 4: Tune hyperparameters using GridSearchCV
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': ['scale', 0.01, 0.1, 1],
    'kernel': ['rbf']
}
grid = GridSearchCV(SVC(), param_grid, cv=5)
grid.fit(X_train, y_train)

print("Best Parameters:", grid.best_params_)
```

```python
# 5.Use cross-validation to evaluate performance
svm_linear_cv_scores = cross_val_score(SVC(kernel='linear', C=100), X_train, y_train, cv=5)
svm_rbf_cv_scores = cross_val_score(SVC(kernel='rbf', C=100, gamma=0.1), X_train, y_train, cv=5)

print("SVM with Linear Kernel Cross-Validation Scores:", svm_linear_cv_scores)
print("SVM with RBF Kernel Cross-Validation Scores:", svm_rbf_cv_scores)
```
